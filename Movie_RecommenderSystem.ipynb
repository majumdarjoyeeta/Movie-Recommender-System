## Import Libraries

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
```

## Load dataset

```python
#loading first dataset
# dataset_reference: https://grouplens.org/datasets/movielens/100k/

movie_titles = pd.read_csv("Movie_title")
movie_titles.head(20)
```

```python
# Let's load the second one!
movies_rating = pd.read_csv('u.data', sep='\t', names=['user_id', 'item_id', 'rating', 'timestamp'])
```

```python
movies_rating.head(10)
```

```python
# Let's drop the timestamp
movies_rating.drop('timestamp', axis=1, inplace=True)
```

```python
movies_rating.head()
```

```python
movies_rating.info()
```

```python
# Let's merge both dataframes together so we can have ID with the movie name
movies_df = pd.merge(movies_rating, movie_titles, on = 'item_id')
```

```python
movies_df.head(20)
```

```python
movies_df.shape
```

## EDA

```python
movies_df.groupby('title')['rating'].describe()
```

```python
ratings_mean_df = movies_df.groupby('title')['rating'].describe()['mean']
```

```python
ratings_count_df = movies_df.groupby('title')['rating'].describe()['count']
```

```python
ratings_count_df
```

```python
ratings_mean_count_df = pd.concat([ratings_count_df, ratings_mean_df], axis = 1)
```

```python
ratings_mean_count_df.reset_index()
```

```python
ratings_mean_count_df['mean'].plot(bins=100, kind='hist', color = 'r')
```

```python
ratings_mean_count_df['count'].plot(bins=100, kind='hist', color = 'r')
```

```python
# Let's see the highest rated movies!

ratings_mean_count_df[ratings_mean_count_df['mean'] == 5]
```

```python
# List all the movies that are most rated
ratings_mean_count_df.sort_values('count', ascending = False).head(100)
```

## Perform Item-Based-Collaborative Filtering on one Movie Sample

```python
userid_movietitle_matrix = movies_df.pivot_table(index = 'user_id', columns = 'title', values = 'rating')
```

```python
userid_movietitle_matrix
```

```python
titanic = userid_movietitle_matrix['Titanic (1997)']
```

```python
titanic
```

```python
# Let's calculate the correlations
titanic_correlations = pd.DataFrame(userid_movietitle_matrix.corrwith(titanic), columns=['Correlation'])
titanic_correlations = titanic_correlations.join(ratings_mean_count_df['count'])
```

```python
titanic_correlations
```

```python
titanic_correlations.dropna(inplace=True)
titanic_correlations
```

```python
# Let's sort the correlations vector
titanic_correlations.sort_values('Correlation', ascending=False)
```

```python
titanic_correlations[titanic_correlations['count']>80].sort_values('Correlation',ascending=False).head()
```
